例外処理の課題完了
*사용자에게 두 개의 숫자를 입력하고 이를 나누는 프로그램을 만듭니다.
*숫자 대신 문자를 입력한 경우 "エラー: 数値を入力してください"라고 표시합니다.
*나누기 수가 0이면 "エラー: 0で割ることはできません"라고 표시합니다.
*계산이 성공적으로 완료되면 결과를 표시합니다.

```python
#例外処理の課題完了
while True:
    try:
        num1 = int(input("1st number: "))
    except ValueError:
        print("エラー: 数値を入力してください")
        continue
    try:
        num2 = int(input("2nd number: "))
        result = num1 / num2
    except ValueError:
        print("エラー: 数値を入力してください")
        continue
    except ZeroDivisionError:
        print("エラー: 0で割ることはできません")
        continue
    else:
        print(f"計算結果: {result}")
        break
```
