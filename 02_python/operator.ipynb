```python
a = 10
b = 3
sum_result = a + b
diff_result = a - b
product_result = a * b
quotient_result = a / b
int_quotient_result = a // b
remainder_result = a % b
power_result = a ** b
print("加算: ", sum_result)          # 13
print("減算: ", diff_result)         # 7
print("乗算: ", product_result)      # 30
print("割算（浮動小数点）: ", quotient_result)  # 3.333...
print("整数除算: ", int_quotient_result)       # 3
print("余り: ", remainder_result)     # 1
print("累乗: ", power_result)
```

```python
x = 10
x += 5  # x = x + 5
print("x += 5: ", x)
x -= 3  # x = x - 3
print("x -= 3: ", x)
x *= 2  # x = x * 2
print("x *= 2: ", x)
x /= 4  # x = x / 4
print("x /= 4: ", x)
x //= 5
print("x //= 5: ", x)
x %= 3
print("x %= 3: ", x)
x **= 3
print("x **= 3: ", x)
```
