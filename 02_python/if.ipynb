```python
score = 80
match score // 10:
    case 10:
        print("満点です！")
    case 9 | 8:
        print("よくできました！")
    case 7 | 6:
        print("合格です！")
    case _:
        print("不合格です")
```

```python
score = 63
match score:
    case score if score == 100:
        print("満点です！")
    case score if score >= 80:
        print("よくできました！")
    case score if score >= 60:
        print("合格です！")
    case _:
        print("赤点です。。。")
```
