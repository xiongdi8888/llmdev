```python
#銀行口座の課題完了
class BankAccount:
    def __init__(self, init_bal = 1000):
        self.__balance = init_bal

    @property
    def balance(self):
        return self.__balance

    def get_balance(self): #USELESS METHOD. .balance is better way. BUT MUST TO USE .get_balance() DUE TO TECHACADEMY SPECIFIED RULE
        print(self.__balance)
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            #print(f"{amount}, {self.__balance}")
        else:
            print("エラー: 正しい金額を入力してください")
    
    def withdraw(self, amount):
        if amount > 0:
            if self.__balance >= amount:
                self.__balance -= amount
                #print(f"{amount}, {self.__balance}")
            else:
                print("エラー: 残高不足または無効な金額です")
        else:
            print("Malicious: Wrong withdrawal amount")

account = BankAccount()
account.deposit(500)
account.withdraw(200)
account.get_balance()
```
