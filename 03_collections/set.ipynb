```python
numbers_list = [1, 2, 3, 2, 1]
unique_numbers = set(numbers_list)
print(unique_numbers)
letters = set("hello")
print(letters)
```

```python
fruits = {"apple", "banana"}
fruits.add("cherry")
print(fruits)
fruits.add("apple")
print(fruits)
```

```python
fruits = {"apple", "banana", "cherry"}
fruits.remove("banana")
print(fruits)
fruits.discard("orange")  # エラーにならない
print(fruits)
removed_item = fruits.pop()
print(fruits)
print(removed_item)
fruits.clear()
print(fruits)
```

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a | b)
print(a & b)
print(a - b)
print(b - a)
print(a ^ b)
```
