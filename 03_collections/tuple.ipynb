```python
numbers = (1, 2, 3, 4, 5)
print(numbers)
fruits = ("apple", "banana", "cherry")
print(fruits)
mixed = (1, "apple", 3.14, True)
print(mixed)
empty_tuple = ()
print(empty_tuple)
one_element_tuple = (1,)
print(one_element_tuple)
```

```python
string_to_tuple = tuple("hello")
print(string_to_tuple)
list_to_tuple = tuple([1, 2, 3])
print(list_to_tuple)
range_to_tuple = tuple(range(5))
print(range_to_tuple)
empty_tuple = tuple()
print(empty_tuple)
```

```python
none_tuple = (None,) * 5
print(none_tuple)
zeros = (0,) * 10
print(zeros)
```

```python
original_list = [1, 2, 3]
copied_tuple = tuple(original_list)
print(copied_tuple)
original_tuple = (4, 5, 6)
copied_tuple2 = tuple(original_tuple)
print(copied_tuple2)
```

```python
fruits = ("apple", "banana", "cherry")
print(fruits[0])
print(fruits[-1])
```

```python
numbers = (0, 1, 2, 3, 4, 5)
print(numbers[1:3])
print(numbers[:3])
print(numbers[2:])
print(numbers[::-1])
```

```python
fruits = ("apple", "banana", "cherry")
print("banana" in fruits)
print(fruits.index("cherry"))
```

```python
numbers = (1, 2, 3, 2, 4, 2, 5)
print(numbers.count(2))
```

```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
combined_tuple = tuple1 + tuple2
print(combined_tuple)
```
