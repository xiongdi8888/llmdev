```python
s = "abcdefghijklmnopqrstuvwxyz"
print(s[0]) # a
print(s[1]) # b
print(s[25])
print(s[-1]) # z
print(s[-3]) # x
print(s[-26])

print(s[26])
print(s[-27])
```

```python
s = "abcdefghijklmnopqrstuvwxyz"
print(s[0:7])
print(s[21:26])
print(s[-5:-1])
```

```python
s = "abcdefghijklmnopqrstuvwxyz"
print(s[:10])
print(s[10:])
print(s[:])
```

```python
s = "abcdefghijklmnopqrstuvwxyz"
print(s[0:10:2])
print(s[1:20:3])
print(s[-1:-10:-2])
```

```python
s = "にんてんどううどんてんに"
s = "Pythonは便利だが遅い"
#s = "no devil lived on"

if s == s[::-1]:
    print(f"「{s}」は回文です！")
else:
    print(f"「{s}」は回文ではないです。")
```

```python
s = "TechAcademy"
for char in s:
    print(char)
```

```python
s = "TechAcademy"
for char in s:
    print(char.upper())
```

```python
s = "TechAcademy"
vowels = "aeiouAEIOU"
for char in s:
    if char in vowels:
        print(char)
```

```python
s = "Hello, world!"
if len(s) > 10:
    print("10文字を超える長い文字列です。")
else:
    print("10文字以内の短い文字列です。")
```
