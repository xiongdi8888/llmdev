```python
#単語の文字数カウントの課題完了
alpha = [_ for _ in "abcdefghijklmnopqrstuvwxyz"]
alpha_hash = dict.fromkeys(alpha, 0)
#print(alpha_hash)
words = []
while True:
    word = input("input words(alpha only): ")
    if word == "":
        break
    elif not word.isalpha() or not word.isascii():
        print("Alphabet only!")
        continue
    elif word in words:
        print("Already exist!")
        continue
    words.append(word)
print(repr(words))
for word in words:
    for char in word:
        if char in alpha_hash:
            alpha_hash[char] += 1
print("文字の出現回数:")
for char, count in {k: v for k, v in alpha_hash.items() if v > 0}.items():
    print(f"{char}: {count}")
```
