#学生の成績分析ツールを作成しよう
* commit comment: 成績分析ツールの課題完了


```python
grades = [85, 90, 78, 92, 88]
n_students = len(grades)
avg_grades = sum(grades) / n_students
best_grade = max(grades)
worst_grade = min(grades)
n_above_avg = sum(1 for grade in grades if grade >= avg_grades)
print(f"平均点: {avg_grades:.1f}\n最高点: {best_grade}\n最低点: {worst_grade}\n平均以上の学生数: {n_above_avg}")
```
